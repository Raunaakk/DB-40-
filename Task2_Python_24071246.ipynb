```python
def connect_database():
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user="root",  
            password="Raunak@9801",  
            database="TRACKER"
        )
        print("Successfully connected to database!")
        return conn
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None
```

```python
def insert_sample_data(connection):
    try:
        cursor = connection.cursor()
        
    
        airlines_data = [
            (1, 'Air Nepal', 'Nepal', 1990, 20),
            (2, 'Air India', 'India', 1932, 150),
            (3, 'Etihad Airway', 'UAE', 2003, 120),
            (4, 'Malaysia Airlines', 'Malaysia', 1947, 80),
            (5, 'Qatar Airways', 'Qatar', 1993, 200)
        ]
        
        cursor.executemany(
            "INSERT INTO AIRLINES VALUES (%s, %s, %s, %s, %s)", 
            airlines_data
        )
        

        flight_data = [
            ('RA230', 'Kathmandu', 'Delhi', 1, 50000, None, 200),
            ('AI215', 'Delhi', 'Kathmandu', 2, 60000, 5000, 300),
            ('3L046', 'Kathmandu', 'Abu Dhabi', 3, 70000, None, 300),
            ('RA416', 'Kathmandu', 'Kuala Lumpur', 1, 55000, 2000, 180),
            ('H9891', 'Kuala Lumpur', 'Kathmandu', 4, 65000, 3000, 250)
        ]
        
        cursor.executemany(
            """INSERT INTO FLIGHT 
            (id, source, destination, aid, fare, discount, capacity) 
            VALUES (%s, %s, %s, %s, %s, %s, %s)""", 
            flight_data
        )
        
        connection.commit()
        print("Successfully inserted 5 rows in each table!")
        
    except mysql.connector.Error as err:
        print(f"Error inserting data: {err}")
        connection.rollback()
    finally:
        if cursor:
            cursor.close()
```

```python
if __name__ == "__main__":
    db_connection = connect_database()
    if db_connection:
        insert_sample_data(db_connection)
        db_connection.close()
        print("Database connection closed")
```
